### 1) Example sparql queries from tutorial

Wikidata:

- same software as Wikipedia
- has items instead of articles
- Wikidata: 122 Mio data entities; Wikipedia (en): 7.2 Mio articles
- Statements:
    - properties and values
        - e.g, properties: "instance of" and value "human"



Sparql:

- query language for linked data databases


```python
## Find all cats on wikidata 

""" 

#Cats
SELECT ?item ?itemLabel #returns a table with two columns
WHERE
{
  ?item wdt:P31 wd:Q146. 
  # Must be a cat; match items against this pattern (); triple: item, property, value; ?item -> any item, i.e. any and all cats on Wikidata
  # Property = P31 = instance of; Value = Q146 (House Cat)
  # all the items that have instance-of with value house cat.
  # wdt & wd (wikidata)= prefixes
  

  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],mul,en". } # Helps get the label in your language, if not, then default for all languages, then en language
}



"""
```

```python
## Find all dogs on wikidata 

""" 
#Dogs
SELECT ?item ?itemLabel
WHERE
{
  ?item wdt:P31 wd:Q144. # Must be a dog
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],mul,en". } # Helps get the label in your language, if not, then default for all languages, then en language
}

"""
```

```python
## Find all American politicians whose father was also a politician 


"""

## List of American politicians whose father was also a politician

# instance of human
# citizenship - US
# occupation - politician
# father - ?father (variable)

SELECT ?item ?itemLabel ?fatherLabel
WHERE {
  ?item wdt:P31 wd:Q5. # instance of human
  ?item wdt:P27 wd:Q30. # citizenship - USA
  ?item  wdt:P106 wd:Q82955. #occupation - politician
  ?item wdt:P22 ?father . #who have *a* father (not optional, because its mandatory)
  ?father wdt:P106 wd:Q82955 . # who *himself* was a politician
  
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],mul,en". } #mul = multiple languages
}

"""
```

```python
## Find all American politicians who were born after 1960 and whose father was also a politician 


""" 

## List of American politicians who were born after 1950 and whose father was also a politician


SELECT ?child ?childLabel ?fatherLabel ?dob
WHERE {
  ?child wdt:P31 wd:Q5. # instance of human
  ?child wdt:P27 wd:Q30. # citizenship - USA
  ?child  wdt:P106 wd:Q82955. #occupation - politician
  ?child wdt:P22 ?father . #who have *a* father (not optional, because its mandatory)
  ?father wdt:P106 wd:Q82955 . # who *himself* was a politician
  
  ?child wdt:P569 ?dob . # child should have *a* date of birth
  FILTER (YEAR(?dob) > 1950) #only allow children born after 1950; function: YEAR
  
  
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],mul,en". }
}

"""
```

### 2) Example sparql query for our use case

- applied to my use case: 
    - what are relevant attributes of the writers (for each: mandatory or optional?), e.g.: 
        - sex or gender
        - date of birth
        - (place of birth)
        - date of death
        - (place of death)
        - writing language (could also use this as a condition instead of country of citizenship)
        - educated at
        - **field of work** -> mandatory or optional
        - **notable work** -> mandatory
        - residence -> OPTIONAL
        - member of 
        - influenced by
        - **award received**

```python
""" 

# 500 German writers OR novelists OR poets with relevant properties 
SELECT ?item ?itemLabel ?sexLabel ?dobLabel ?pobLabel ?dodLabel ?podLabel ?lowLabel ?educLabel ?fowLabel ?nwLabel ?resLabel ?memberLabel ?inflLabel ?awardLabel ?movemLabel
WHERE {
  ?item wdt:P31 wd:Q5. # humans
  
  # German citizenship OR writes in German language 
  {?item wdt:P27 wd:Q183.} # citizenship - Germany
  UNION
  {?item wdt:P6886 wd:Q188.} # language written - German
  
  # occupation: writer OR novelist OR poet                                  
  {?item  wdt:P106 wd:Q36180.} #occupation - writers
  UNION
  {?item  wdt:P106 wd:Q6625963 .} #occupation - novelist
  UNION
  {?item wdt:P106 wd:Q49757 .} # occupation - poet
  
  
  OPTIONAL {?item wdt:P21 ?sex }. #put the sex/gender into a variable
  OPTIONAL {?item wdt:P569 ?dob} .#put the date of birth into a variable
  OPTIONAL {?item wdt:P19 ?pob }. # put the place of birth into a variable
  OPTIONAL {?item wdt:P570 ?dod }. #put the date of death into a variable; make it optional (some authors are still alive)
  OPTIONAL {?item wdt:P20 ?pod }. #put the place of death into a variable
  OPTIONAL {?item wdt:P6886 ?low }. #put the language of writing into a variable
  OPTIONAL {?item wdt:P69 ?educ }. #put the education institutions into a variable
  OPTIONAL {?item wdt:P101 ?fow }. #put the field of work into a variable
  OPTIONAL {?item wdt:P800 ?nw }. #put the notable into a variable
  OPTIONAL {?item wdt:P551 ?res} . #put the residence into a variable; make it optional
  OPTIONAL {?item wdt:P463 ?member} . #put the membership into a variable; make it optional
  OPTIONAL {?item wdt:P737 ?infl } . #put the influence into a variable; make it optional
  OPTIONAL {?item wdt:P166 ?award } . #put the award into a variable; make it optional
  OPTIONAL {?item wdt:P135 ?movem } . #put the movement into a variable; make it optional
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],mul,en". }
}
LIMIT 500

"""
```
